# Extrair a série de dados do Brasil (loc())

loc[ ]- pesquisa pelo label
iloc[ ] - pesquisa pelo indice

```python
# Extrair a serie de dados do Brasil

brasil = df.loc['Brasil',anos]
```

- criando um dicionário

```python
brasil_dict = {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
```

```python
brasil_dict
```

- transformar o brasil em um dataframe

```python
df_brasil = pd.DataFrame(brasil_dict)
```

```python
df_brasil
```

# Criando o gráfico de linhas

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.grid()
plt.show()
```

# Transformando um gráfico como figura
- aumenta o potencial dos comandos


```python


# 8 - polegadas
# 4 - polegadas
# # 1 polegada = 2.54 cm
# largura 8 = 20.32 cm
# altura  4 = 10.16 cm
# DPI (dots per inch, ou pontos por polegada).
# Padrão no matplotib - DPI = 100
# Largura : 8 polegadas x 100 DPI = 800 pixels.
# Altura  : 4 polegadas x 100 DPI = 400 pixels.
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano',fontsize=14)
plt.ylabel('Numero de Imigrantes',fontsize=14)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.figure(figsize=(8,4))
plt.show()
```

# Criando subplots e adicionando rótulos

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(df_brasil['ano'],df_brasil['imigrantes'],lw=4,color='#2E8FF2',marker='o')
ax.set_title("Imigração do Brasil para o Canadá \n 1980 a 2013",loc="left")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_xlabel('Ano')
ax.set_ylabel('Numero de imigrantes')
plt.grid(linestyle='--')
plt.show()
```

# Gráfico de linha e boxplot em subplots

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))
# subplot 1 grafico de linha

axs[0].plot(df_brasil['ano'],df_brasil['imigrantes'])
axs[0].set_title("Imigração do Brasil para o Canadá \n 1980 a 2013")
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Numero de imigrantes')
axs[0].grid()
# subplot 2 grafico boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title("Box-plot da imigração do Brasil para o Canadá\n 1980 a 2013")
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Numero de imigrantes')
axs[1].grid()
```

### Comparativo dos maiores paises da America Latina (imigração Canadá)

```python
# Criar uma figura

fig, axs = plt.subplots(2,2,figsize=(10,6))

axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('Brasil')

#Colombia
axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Colômbia')

#Argentina
axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

#Peru
axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')

fig.subplots_adjust(hspace=0.5,wspace=0.3)

for ax in axs.flat: #uso do array bidimensional
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
for ax in axs.flat:
    ax.set_xlabel("Ano")
    ax.set_ylabel("Número de imigrantes")

    # Usando ravel() - achatar a mtriz evitando o uso complexo do fo 
    # Usando set_ylim() - definir o limite os valores
    # determinar o minimo 

    ymin = 0
    # determinar o maximo
    ymax = 7000
    # criar um for usar ravel() - achatar a matriz  bi para uma uni
    for ax in axs.ravel(): #uso do array unidimensional
        #usar set_ylim() - determinar o mínimo e o máximo do eixo y
        ax.set_ylim(ymin,ymax)

    #suptitle
    fig.suptitle("Imigração na América Latina",fontsize=16)
    #Customizando o Matplotlib (cores, fontes, bordas)


```

### Exercicio - Verifique na documentação do Matplotlib
- função ravel()
- função set_ylim()

função ravel
Em muitos casos ao trabalhar com Matplotlib, você pode ter arrays multidimensionais (como uma matriz de valores para uma imagem) e pode ser necessário achatar esses arrays para algumas operações, como quando você precisa de uma lista unidimensional de valores para gerar um histograma ou ajustar alguns gráficos.

função set_ylim()
A função set_ylim no Matplotlib é usada para definir os limites do eixo y de um gráfico. Isso é útil para ajustar a escala vertical do gráfico e garantir que os dados sejam apresentados de forma clara e eficaz.

# Grafico de Barras

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(america_sul.index,america_sul['Total'],color=cores)
ax.set_title('Imigração da América do Sul para o Canadá')
ax.set_xlabel('')
ax.set_ylabel('Número de Imigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=9)
ax.yaxis.set_tick_params(labelsize=9)
plt.show()
```

## Deixando o gráfico na horizontal

- criar um gráfico invertido
- de vertical para horizontal

```python
#criando uma variavel fig para a figura
#ax é a variavel que representa o grafico (eixos e valores... )
fig, ax = plt.subplots(figsize=(12,5))
# é um grafico de bar(matplotlib)
ax.barh(america_sul.index,america_sul['Total'],color=cores)
# titulo alinha a esquerda com fonte 18
ax.set_title('Imigração da América do Sul para o Canadá')
# eixo y com fonte 14
ax.set_xlabel('')
ax.set_ylabel('Número de Imigrantes',fontsize=14)
# mudar as fontes dos ticks
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)
plt.show()
```

```python
america_sul_sorted = america_sul.sort_values(by="Total", ascending=True)
```

```python
america_sul_sorted.head()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index,america_sul_sorted['Total'],color=cores)
ax.set_title('Imigração da América do Sul para o Canadá')
ax.set_xlabel('')
ax.set_ylabel('Número de Imigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)
plt.show()
```
